# WEB SCRAPING
The technique of taking the html file sent by the server into python and scrapping it instead of giving it to the browser and displaying it is called Web scrapping.

Two ways of getting data from a website:

1. Using API
2. HTML web scrapping using some tool like bs4

```python
import requests
from bs4 import BeautifulSoup
```

### Requesting the site from URL

```python
url = "https://codewithharry.com"
r = requests.get(url) # r variable has all the HTML code
```

### HTML content calling and then parsing it

```python
htmlContent = r.content # r returns response so if we want the code we write r.content
soup = BeautifulSoup(htmlContent, 'html.parser') # use beautiful soup for parsing html content
print(htmlContent) # printing the code
print("\n \n \n \n")
print(soup)
print("\n \n \n \n")
print(soup.prettify()) 
```

### HTML Tree Traversal:

```python
# Get the Title
title = soup.title
print(title)
```

```python
# Get the paragraph
print(soup.find('p')) # find the first paragraph tag
print('\n \n \n \n')
paras = soup.find_all('p') # find all the paragraphs tag
print(paras)
```

```python
# Get the anchors
print(soup.find('a')) # find the first anchor tag
print('\n \n \n \n')
anchors = soup.find_all('a') # find all the anchors tag
print(anchors)
```

```python
# Get classes and ids of any element in the HTML page 
print(soup.find('p')['class'])
print(soup.find('div')['id'])
```

```python
# find all the elements with class lead 
print(soup.find_all("p", class_="lead"))
```

```python
# Get the text from the tags/soup 
print (soup.find('p').get_text()) 
print('\n \n \n \n')
print(soup.get_text())
```

```python
# Get all the links
soup.find('a').get('href') # get a single link


# get all the links

link = soup.find_all('a')

for i in link:
    print(i.get('href'))
    
    
# get all the unique links

all_links = set()

link = soup.find_all('a')
count=0
for i in link:
    if(i.get('href')!='#'):
        link =  "https://codewithharry.com" + i.get('href')
        all_links.add(link)
        print(link)
        count=count+1
print(count)
print('\n \n \n')
print(all_links)
print(len(all_links))
```

```python
# find an id and seacrh along its contents and children
# (according to DOM tree structure)
imgpreview2 = soup.find(id = "imgpreview2")
print(imgpreview2.contents)
```

```python
# children of that id
for ele in imgpreview2.children:
    print(ele.name)
for ele in imgpreview2.children:
    print(ele)
    print("\n \n")
```

```python
# parent of that id
for ele in imgpreview2.parents:
    print(ele.name)
for ele in imgpreview2.parents:
    print(ele)
    print("\n \n")
```

```python
print(imgpreview2.next_sibling.next_sibling.name)
print(imgpreview2.previous_sibling.previous_sibling)
```

```python
# strings and stripped strings
for ele in imgpreview2.strings:
    print(ele)
print('\n\n')
for ele in imgpreview2.stripped_strings:
    print(ele)
```

```python
# to extract a class using '.' and '#'
work = soup.select('.login-modal')
print(work)
```

## TRADE SITE SCRAPE

### Extracting the columns first and then extracting the data

```python
url = "https://ticker.finology.in/"
r= requests.get(url)
soup = BeautifulSoup(r.text,"lxml")
```

```python
table = soup.find("table", class_ ="table table-sm table-hover screenertable") 
headers = table.find_all("th") 
print (headers)
```

```python
titles =[]
for i in headers:
    title = i.text
    titles.append(title)
titles
```

```python
trade_df = pd.DataFrame(columns = titles)
trade_df
```

```python
rows = table.find_all("tr")
rows
```

```python
for i in rows[1:]:
    data = i.find_all("td")
#     print(data)
    row =  [tr.text for tr in data]
#     print(row)

    l = len(trade_df)

    trade_df.loc[l]=row
trade_df
```

```python
trade_df.to_csv("trading.csv")
```

### Extracting details from multiple pages

```python
i=1
def page_change(i):
    url = "https://www.flipkart.com/search?q=mobiles+under+50000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=" + str(i)
    r=requests.get(url)
    # print(r)
    soup = BeautifulSoup(r.text,"lxml")
    # print(soup)
    box = soup.find("div" ,class_ = "_1YokD2 _3Mn1Gg")
    # print(box)
```

```python
product_name = []
product_price = []
product_desc = []
product_review = []
```

```python
def naming():
    names = box.find_all ("div" , class_ = "_4rR01T") 
    for i in names:
        name = i.text
        product_name.append(name)
    product_name
    # len(product_name)
```

```python
def pricing():
    prices = box.find_all("div", class_="_30jeq3 _1_WHN1") 
    for i in prices:
        price = i.text
        product_price.append(price)
    product_price
    # len(product_price)
```

```python
def describing():
    description = box.find_all ("ul" , class_ = "_1xgFaf") 
    for i in description:
        desc = i.text
        product_desc.append(desc)
    product_desc
    # len(product_desc)
```

```python
def reviewing():
    reviews = box.find_all ("div" , class_ = "_3LWZlK") 
    for i in reviews:
        rev = i.text
        product_review.append(rev)
    product_review
    # len(product_review)
```

```python
for i in range(1,11):
    page_change(i)
    naming()
    pricing()
    describing()
    reviewing()
```

```python
Phone_df= pd.DataFrame({"product_name" : product_name , "product_price" : product_price , "product_desc" : product_desc , "product_review" : product_review})
Phone_df
```

```python
# driver.get("https://www.airbnb.co.in/?check-account-activation=1")
driver.get("https://www.airbnb.co.in/s/North-Goa--Goa/homes?tab_id=home_tab&refinement_paths%5B%5D=%2Fhomes&flexible_trip_lengths%5B%5D=one_week&monthly_start_date=2023-08-01&monthly_length=3&price_filter_input_type=0&price_filter_num_nights=5&channel=EXPLORE&query=North%20Goa%2C%20Goa&place_id=ChIJMaD6Uu3AvzsR0E3VfowKzRs&date_picker_type=calendar&checkin=2023-08-17&checkout=2023-08-18&adults=6&source=structured_search_input_header&search_type=autocomplete_click")
```

```python
names = []
prices = []
descriptions = []
```

```python
## EXTRACT THE NAMES
def extract():
    # Assuming you have already initialized the WebDriver instance as 'driver'
    property_names = driver.find_elements_by_class_name("t1jojoys")

    for i in property_names:
        # Get the text content of the element
        prop = i.text
        # Print the extracted text
        names.append(prop)
#         print(prop)




    ## EXTRACT THE PRICE

    property_prices = driver.find_elements_by_class_name("_tt122m")


    for i in property_prices:
        # Get the text content of the element
        price = i.text
        # Print the extracted text
        prices.append(price)
#         print(price)




    ## EXTRACT THE DESCRIPTION

    property_desc = driver.find_elements_by_class_name("t6mzqp7")


    for i in property_desc:
        # Get the text content of the element
        desc = i.text
        # Print the extracted text
        descriptions.append(desc)
#         print(desc)
```

```python
# Assuming you have already initialized the WebDriver instance as 'driver'
length = int(driver.find_element_by_xpath("/html/body/div[5]/div/div/div[1]/div/div[2]/div/div/div/div[1]/div[2]/main/div[2]/div/div[3]/div/div/div/nav/div/a[4]").text)
# print(length)
for i in range(length):
    extract()
    time.sleep(2)
    element = driver.find_element_by_xpath("/html/body/div[5]/div/div/div[1]/div/div[2]/div/div/div/div[1]/div[2]/main/div[2]/div/div[3]/div/div/div/nav/div/a[5]").click()
# print(element)
```

```python
property_df= pd.DataFrame({"property_name" : names , "property_price" : prices , "property_desc" : descriptions })
property_df
```
